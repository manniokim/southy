```python
gender_series = foodie['Gender'].value_counts()
labels = (np.array(gender_series.index))
sizes = (np.array((gender_series /gender_series.sum())*100))

trace = go.Pie(labels=labels, values=sizes)
layout = go.Layout(
    title='Gender distribution'
)
data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="gender")
```
